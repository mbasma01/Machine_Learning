```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
# Load the Iris dataset from sklearn
iris = load_iris()
```

```python
# Create a Pandas DataFrame from the data
data = iris.data
feature_names = iris.feature_names
target = iris.target
target_names = iris.target_names

df = pd.DataFrame(data, columns=feature_names)
df['species'] = [target_names[i] for i in target]
```

```python
#an alternative using iris.csv
#df = pd.read_csv('iris.csv')
```

```python
# Display the first 5 rows
print("First 5 rows:")
print(df.head().to_markdown(index=False, numalign="left", stralign="left"))
```

```python
# Get information about the data (columns, data types, non-null values)
print("\nData Information:")
print(df.info())

# Summary statistics
print("\nSummary Statistics:")
print(df.describe().to_markdown(numalign="left", stralign="left"))

# Selecting specific columns
print("\nSepal Length and Width:")
print(df[['sepal length (cm)', 'sepal width (cm)']].head().to_markdown(index=False, numalign="left", stralign="left"))

# Filtering data
print("\nIris Versicolor data:")
print(df[df['species'] == 'Iris-versicolor'].head().to_markdown(index=False, numalign="left", stralign="left"))

# Sorting
print("\nSorted by Petal Length (descending):")
print(df.sort_values(by='petal length (cm)', ascending=False).head().to_markdown(index=False, numalign="left", stralign="left"))

# Simple Visualization
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'], c=df['species'].map({'Iris-setosa': 'blue', 'Iris-versicolor': 'orange', 'Iris-virginica': 'green'}))
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Length vs. Sepal Width')
plt.show()
```
