```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset from the CSV file
df = pd.read_csv("boston_house_prices.csv")
```

```python
# Display the first few rows and get a summary of the dataset
print("First 5 rows:")
print(df.head().to_markdown(index=False, numalign="left", stralign="left"))  # Use to_markdown for better formatting

print("\nDataset Information:")
print(df.info())
```

```python
# Check for missing values
print("\nMissing Values:")
df.isnull().sum()
```

```python
# Descriptive statistics for numerical columns
print("\nDescriptive Statistics:")
df.describe()
```

```python
# Distribution of houses near the Charles River
print("\nHouses Near Charles River (CHAS=1):")
print(df['CHAS'].value_counts().to_markdown(numalign="left", stralign="left"))
```

```python
#visualising data distribution by age
age_count = round(df['AGE']).value_counts() #counting rows of distinct ages
#plotting the distributed point (x, y);
#x is the age, y is the frequency of the age
plt.scatter(age_count.index, age_count.values) 
plt.xlabel("AGE")
plt.ylabel("count")
plt.title("Distribution of houses by age")
```

```python
#filtering out crime rate outliers
lower_crime = df[df['CRIM'] < 25]
```

```python
#visualising crime rate distribution using a histograms
lower_crime_count = round(lower_crime['CRIM'], 1).value_counts()
plt.hist(lower_crime_count)
plt.show()
plt.close()
```

```python
# Correlation matrix
print("\nCorrelation Matrix:")
df.corr()
```

```python
import seaborn as sns   # visualization
```

```python
plt.figure(figsize=(14,12))
sns.heatmap(df.corr(),annot=True)
plt.xticks(rotation=45)
plt.show()
plt.close()
```

```python
plt.scatter(df['RM'],df['MEDV'])
```

```python
# Visualizations

# Scatter plot of crime rate vs. median house value
plt.figure(figsize=(10, 6))  # Adjust figure size for better readability
plt.scatter(df['CRIM'], df['MEDV'])
plt.xlabel('Crime Rate per Capita')
plt.ylabel('Median House Value ($1000s)')
plt.title('Relationship between Crime Rate and House Value')
plt.grid(True)  # Add a grid for better visualization
plt.show()
```

```python
# Histogram of average number of rooms
plt.figure(figsize=(10, 6))
plt.hist(df['RM'], bins=10, edgecolor='black')  # Adjust number of bins and add edge color
plt.xlabel('Average Number of Rooms per Dwelling')
plt.ylabel('Frequency')
plt.title('Distribution of Average Number of Rooms')
plt.show()
```

```python
# Box plot comparing median house values by proximity to Charles River
plt.figure(figsize=(8, 5))
sns.boxplot(x='CHAS', y='MEDV', data=df)
plt.xlabel('Proximity to Charles River (1=Near, 0=Not Near)')
plt.ylabel('Median House Value ($1000s)')
plt.title('House Values by Proximity to Charles River')
plt.show()
```
