```python
x = df.drop('MEDV', axis=1)
```

```python
y = df['MEDV']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.15)
```

```python
x_test
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
len(y_pred)
```

```python
plt.scatter(x_test['LSTAT'], y_test, color = 'red')
plt.scatter(x_test['LSTAT'], y_pred, color = 'blue')
```

```python
X = df[['RM', 'LSTAT']]
Y = df['MEDV']
model_basic = LinearRegression()
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.15)
```

```python
model_basic.fit(X_train, Y_train)
```

```python
Y_pred = model_basic.predict(X_test)
plt.scatter(X_test['RM'], Y_test, color = 'red')
plt.scatter(X_test['RM'], Y_pred, color = 'blue')
```
